Zaimportowanie pakietu SWAT i utworzenie połączenia do CAS

```python
import swat
```

```python
conn = swat.CAS('cas-controller', 5570)
```

```python
conn.serverStatus()
```

```python
conn.sessionStatus()
```

Akcja **help**.

```python
conn.help()
```

```python
out
```

## Uruchamianie CAS Actions

Akcja **userinfo**

```python
conn.userinfo()
```

Aby ładniej wyświetlać rezultaty można użyc modulu **pprint** (pretty print), który wyswietla rezultaty jako słownik.

```python
from pprint import pprint

pprint(dict(conn.userinfo()))
```

Wywołanie akcji **userinfo** uzywając pełnej ścieżki i nazwy.

```python
conn.builtins.userinfo()
```

## Ładowanie danych

Metoda **upload** obiektu połączenia **CAS** ładuje dane lokalne do pamięci serwera CAS.

```python
out = conn.upload('../../dane/iris.csv')
out
```

Uchwyt do tabeli utworzonej poprzez załadowanie pliku iris.csv. Obiekt **CASTable** odwołuje się do tabeli CAS powstałej po stronie serwera.

```python
iris = out.casTable
```

Metoda  **CASTable.tableinfo** zwraca informacje o tabeli.

```python
iris.tableinfo()
```

Metoda **CASTable.columninfo** analogicznie - informacje o kolumnach.

```python
iris.columninfo()
```

## Wykonywanie akcji na tabeli CAS

Z poziomu IPython można podejrzeć pomoc dla akcji bądź zestawu akcji **CAS**

```python
conn.simple?
```

Metoda **CASTable.summary**

```python
summ = iris.summary()
summ
```

Metoda **CASTable.describe**. Ta metoda emuluje metodę **describe** z pakietu Pandas, ale wywołuje akcję CAS do wyliczenia resultatów.

```python
iris.describe()
```

## Wizualizacja danych

Metoda **CASTable.plot**

```python
iris.plot()
```

Wykres może nie wyświetlić się automatycznie. Zalezy to od Twojej konfiguracji. Możesz użyć metody **show** do wyświetlenie wykresu.

```python
import matplotlib.pyplot as plt
```

```python
plt.show()
```

Alternatywnie, można użyć magii **%matplotlib** w Jupyter notebook aby wyswietlić wykresy w sposób inline.

```python
%matplotlib inline
```

```python
iris.plot()
```

Sortowanie danych do wykresu.

```python
iris.sort_values(['sepal_length', 'sepal_width']).plot()
```

Zamknięcie połączenia.

```python
conn.close()
```

Zamknięcie sesji.

```python
conn.endsession()
```
