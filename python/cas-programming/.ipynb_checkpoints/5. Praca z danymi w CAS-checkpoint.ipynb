## Żonglowanie danymi w locie

### Tworzenie kolumn wyliczanych

```python
sorttbl['sepal_factor'] = ((sorttbl.sepal_length + sorttbl.sepal_width) * 2)
sorttbl.head()
```

```python
sorttbl['total_factor'] = sorttbl.sepal_factor + sorttbl.petal_width + sorttbl.petal_length
sorttbl.head()
```

```python
sorttbl['names'] = 'sepal / petal'
sorttbl.head()
```

```python
sorttbl['cap_names'] = sorttbl.names.str.title()
sorttbl.head()
```

```python
sorttbl
```

### Przetwarzanie w grupach

Specyfikacja gryp bezpośrednio na obiekcie CASTable.

```python
tbl.set_param('groupby', ['variety'])
tbl
```

```python
tbl.summary(subset=['min', 'max'])
```

```python
tbl.del_param('groupby')
tbl
```

Użycie metod DataFrame do ustawienia grupowania.

```python
grptbl = tbl.groupby(['variety'])
grptbl
```

```python
grptbl.summary(subset=['min', 'max'])
```

#### Konkatenacja zbiorów z grupowaniem By Groups

```python
grpsumm = grptbl.summary(subset=['min', 'max'])
grpsumm
```

```python
grpsumm.concat_bygroups()
```

#### Wybieranie rezultatów po nazwie tabeli

```python
grpcorr = grptbl.correlation()
grpcorr
```

```python
grpcorr.get_tables('Correlation')
```

```python
swat.concat(grpcorr.get_tables('Correlation'))
```

#### Wybeiranie konkretnyj grupy

```python
grpsumm.get_group(['Versicolor'])
```

```python
grpsumm.get_group(variety='Versicolor')
```

#### Grupowanie na kilka sposobów na raz

```python
grpmdsumm = tbl.mdsummary(sets=[dict(groupby=['sepal_length']),
                                dict(groupby=['petal_length'])])
```

```python
list(grpmdsumm.keys())
```

```python
grpmdsumm.get_set(1)
```

```python
grpmdsumm.get_set(1).concat_bygroups()
```
